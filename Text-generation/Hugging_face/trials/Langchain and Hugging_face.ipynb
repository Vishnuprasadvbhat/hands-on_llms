# HuggingFace Endpoint
## Calling huggingFace Models with API

### Two major ways of Calling A Model:
### 1. Download the model to the local --> requires RAM 
### 2. Calling models through Endpoints




```python
from langchain_huggingface import HuggingFaceEndpoint
```

```python
import os 
os.environ['HUGGINGFACE_API_TOKEN'] = Hf_token
```

```python
repo_id= "mistralai/Mistral-7B-Instruct-v0.3"

llm = HuggingFaceEndpoint(repo_id=repo_id,max_new_tokens=128,temperature=0.7, huggingfacehub_api_token="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")
```

```python
llm
```

```python
llm.invoke("What is machine learning?")
```

```python
from langchain import PromptTemplate , LLMChain

question = "when will RCB win the trophy"

template= """ Question: {question}
Answer: let's think step by Step. """

prompt = PromptTemplate(template=template, input_variables=['question'])
print(prompt)
```

```python
llm_chain= LLMChain(llm=llm, prompt=prompt)
print(llm_chain.invoke(question))
```

## When we download the entire Model to our local system  we use Huggingface Pipeline than calling the model using huggingface inference Model

Dis: Some LLMs are huge, cannot be downloaded

```python
from langchain_huggingface import HuggingFacePipeline
```

```python
# using Tranformers we can call any model, and each model has its own tokenizer
from transformers import AutoModelForCausalLM, AutoTokenizer, pipeline
```

```python
model_id = "gpt2"
# model_id ="gpt2-xl"
AutoModelForCausalLM.from_pretrained(model_id)
tokenizer= AutoTokenizer.from_pretrained(model_id)
```

```python
pipe= pipeline('text-generation',model=model_id, tokenizer=tokenizer,token=1024)

hfg= HuggingFacePipeline(pipeline=pipe)
```

```python
hfg.invoke("I am superhero")
```
